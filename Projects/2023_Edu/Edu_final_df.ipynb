```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid', font_scale=1.5)
sns.set_palette('Set2', n_colors=10)
plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
```

```python
df = pd.read_excel('우혁 데이터\청소년 현재 흡연율 추이.xlsx')
df = df.iloc[3:6]
df.columns = ['학교 구분','2019','남학생','여학생','2020','남학생','여학생','2021','남학생','여학생','2022','남학생','여학생']
df = df.loc[:,['학교 구분','2019','2020','2021','2022']]
df = df.set_index('학교 구분').T
df.reset_index(inplace=True)
df['연도'] = pd.to_numeric(df['index'])
df['전체'] = pd.to_numeric(df['전체'])
df['중학교'] = pd.to_numeric(df['중학교'])
df['고등학교'] = pd.to_numeric(df['고등학교'])
```

```python
fig, ax = plt.subplots()
ax.plot(df['연도'], df['전체']); 
```

```python
df = pd.read_excel('우혁 데이터\코로나19 이후 학업 스트레스 변화.xlsx')
df
```

```python
df.columns= ['기타','연령/성별','매우 감소하였다','약간 감소하였다','변화 없다','약간 증가하였다','매우 증가하였다','평균']
df = df.iloc[1:,1:]
df.reset_index(drop = True)
df['매우 감소하였다'] = pd.to_numeric(df['매우 감소하였다'])
df['약간 감소하였다'] = pd.to_numeric(df['약간 감소하였다'])
df['변화 없다'] = pd.to_numeric(df['변화 없다'])
df['약간 증가하였다'] = pd.to_numeric(df['약간 증가하였다'])
df['매우 증가하였다'] = pd.to_numeric(df['매우 증가하였다'])
df['평균'] = pd.to_numeric(df['평균'])
df_age = df.iloc[:3]
df_sex = df.iloc[3:]
```

```python
df_age = df_age.set_index('연령/성별').T
df_sex = df_sex.set_index('연령/성별').T
```

```python
fig, ax = plt.subplots()
ax.plot(df_age['만9-12세'],data = df_age)
plt.xticks(rotation = 45)
plt.title('만9-12세 코로나 이후 스트레스');
```

```python
fig, ax = plt.subplots()
ax.plot(df_age['만13-18세'],data = df_age)
plt.xticks(rotation = 45)
plt.title('만13-18세 코로나 이후 스트레스');
```

```python
fig, ax = plt.subplots()
ax.plot(df_age['만19-24세'],data = df_age)
plt.xticks(rotation = 45)
plt.title('만19-24세 코로나 이후 스트레스');
```

```python
fig, ax = plt.subplots()
ax.plot(df_age['만9-12세'],data = df_age)
ax.plot(df_age['만13-18세'],data = df_age)
ax.plot(df_age['만19-24세'],data = df_age)
plt.xticks(rotation = 45)
plt.title('만19-24세 코로나 이후 스트레스');
plt.legend(loc=2, bbox_to_anchor=(1,1));
```

```python
fig, ax = plt.subplots()
ax.plot(df_sex['남'],data = df_sex)
ax.plot(df_sex['여'],data = df_sex)
plt.xticks(rotation = 45)
plt.title('성별 코로나 이후 스트레스');
plt.legend(loc=2, bbox_to_anchor=(1,1));
```
